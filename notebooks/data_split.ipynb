# Data Split

We will split the data to avoid data leakage by separating training, test and holdout sets by years.

```python
import pandas as pd
```

```python
df = pd.read_csv('/home/heisenberg_0702/Housing-Pricing-Regression/data/raw/raw_housing_data.csv')
```

```python
df['date'] = pd.to_datetime(df['date'])
print(df['date'].dt.year.value_counts()) # Print the count of records per year
```

**To avoid leakage**

We will split the data in a time-based fashion. 

- Train set: (2012-2019)
- Test set: (2020-2021)
- Holdout set: (2022-2023)

```python
# Sort the dataframe by date
df = df.sort_values(by='date')
```

```python
# Define the cut-off date for splitting
cutoff_date_test = "2020-01-01" # Testing start date
cutoff_date_holdout = "2022-01-01" # Validation start date
```

```python
# Train set: data before 2020-01-01
train_df = df[df['date'] < cutoff_date_test]

# Test set: data from 2020-01-01 to 2021-12-31
test_df = df[(df['date'] >= cutoff_date_test) & (df['date'] < cutoff_date_holdout)]

# Holdout set: data from 2022-01-01 onwards
holdout_df = df[df['date'] >= cutoff_date_holdout]

# Shapes of the splits
print("Train set shape:", train_df.shape)
print("Test set shape:", test_df.shape)
print("Holdout set shape:", holdout_df.shape)
```
